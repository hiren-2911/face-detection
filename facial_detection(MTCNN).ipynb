```
#importing libraries 
from mtcnn.mtcnn import MTCNN
import matplotlib as mlt
import matplotlib.pyplot as plt
```

```
filename='/content/drive/My Drive/facial_detect.jpeg'#file path
pixels=plt.imread(filename)#reading the file
detector = MTCNN()#instanciating the detector object
faces=detector.detect_faces(pixels)#calling the detect_faces method on detector object 
#for face in faces:
#	print(face)
result=faces#it returns a list contating the x,y,width and height.it also contains the dictonary contaiting landmarks 
len(result)
#result
```

```
#making bounding box on the face and marking the landmarks
for i in range(len(faces)):
  data = plt.imread(filename)
  plt.imshow(data)
  x,y,width,height=faces[i]['box']
  ax=plt.gca()#getting the present axis on which the image is plotted
  rect=mlt.patches.Rectangle((x, y), width, height, fill=False, color='red')#marking rectangle
  plt.rcParams['lines.markersize']=5 #setting markersize small so it dosent overlap
  ax.scatter(result[i]['keypoints']['left_eye'][0],result[i]['keypoints']['left_eye'][1],color='r')
  ax.scatter(result[i]['keypoints']['right_eye'][0],result[i]['keypoints']['right_eye'][1],color='r')
  ax.scatter(result[i]['keypoints']['nose'][0],result[i]['keypoints']['nose'][1],color='r')
  ax.scatter(result[i]['keypoints']['mouth_left'][0],result[i]['keypoints']['mouth_left'][1],color='r')
  ax.scatter(result[i]['keypoints']['mouth_right'][0],result[i]['keypoints']['mouth_right'][1],color='r')
  ax.add_patch(rect)
plt.show()

```
